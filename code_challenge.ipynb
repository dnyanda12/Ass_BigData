import matplotlib.pyplot as plt: Imports the matplotlib.pyplot module for creating visualizations and assigns it the alias plt

from sklearn.preprocessing import StandardScaler: Imports StandardScaler from the scikit-learn library to normalize the data. This is an important step to ensure features with different scales (like population and density) are treated equally by the clustering algorithm.

from sklearn.cluster import KMeans: Imports the KMeans algorithm from the scikit-learn library. This is the unsupervised machine learning algorithm that performs the clustering.

X = world_pop[['2022 Population','2015 Population', '2010 Population','Density (per km²)']].dropna():

This selects the specified columns from the world_pop GeoDataFrame. These columns represent the features that will be used for clustering.
.dropna() removes any rows with missing data (NaN) in any of the selected columns. This is a necessary step, as the K-Means algorithm cannot handle missing values.

scaler = StandardScaler(): Creates an instance of the StandardScaler object.

X_scaled = scaler.fit_transform(X):
fit_transform() is a method that first "fits" the scaler to the data in X (calculates the mean and standard deviation for each column) and then "transforms" the data by standardizing it.
This ensures all features have a mean of 0 and a standard deviation of 1. If this step were skipped, features with a larger scale (like population) could disproportionately influence the distance calculations and skew the clustering results.


kmeans = KMeans(n_clusters=5, random_state=42):
Creates an instance of the KMeans algorithm.

n_clusters=5 tells the algorithm to group the data into 5 clusters. The number of clusters, k, must be chosen manually.

random_state=42 ensures that the results are reproducible. Without it, the algorithm would randomly initialize the starting points of the clusters, and each run would likely produce slightly different results.

world_pop['cluster'] = kmeans.fit_predict(X_scaled):
fit_predict() trains the K-Means algorithm on the normalized data (X_scaled) and then predicts which cluster each country belongs to.
The resulting cluster labels (e.g., 0, 1, 2, 3, 4) are added as a new column named 'cluster' to the original world_pop GeoDataFrame.


world_pop.plot(column='cluster', legend=True, figsize=(12, 8)):
This uses GeoPandas' plotting functionality to create a map.

column='cluster' tells the function to color each country according to its assigned cluster.

legend=True displays a legend to the side of the map.

figsize=(12, 8) sets the size of the plot for better readability.
plt.title("Country Clusters Based on Population and Area"): Adds a descriptive title to the map.

plt.show(): Displays the final plot. The resulting map will visually show the five country clusters based on their population trends and density.

```python
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Select features
X = world_pop[['2022 Population','2015 Population', '2010 Population','Density (per km²)']].dropna()

# Normalize
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply KMeans
kmeans = KMeans(n_clusters=5, random_state=42)
world_pop['cluster'] = kmeans.fit_predict(X_scaled)

# Visualize clusters
world_pop.plot(column='cluster', legend=True, figsize=(12, 8))
plt.title("Country Clusters Based on Population and Area")
plt.show()
```


import geopandas as gpd and import pandas as pd: These lines import the necessary libraries. GeoPandas (gpd) is for handling the geographic data, while Pandas (pd) is used for the tabular population data.
shapefile_path and csv_path: These variables store the file paths as strings for readability.


gdf = gpd.read_file(...): This command uses GeoPandas to read the world administrative boundaries shapefile into a GeoDataFrame named gdf. A GeoDataFrame is essentially a Pandas DataFrame with a special column for geometric shapes (the country borders)

df = pd.read_csv(...): This command uses Pandas to read the world population CSV file into a standard DataFrame named df.

print(df.columns) and print(gdf.columns): These lines are diagnostic steps. They print the column names of both DataFrames, which is crucial for identifying the correct columns to use for the join.

gdf = gdf.loc[:,~gdf.columns.duplicated()]: This is a data cleaning step. It removes any columns from the gdf GeoDataFrame that have duplicate names, keeping only the first occurrence of each unique column name. This is important to prevent unexpected behavior during the merge.


gdf = gdf.merge(...): This is the core of the code. It merges the gdf GeoDataFrame with the df DataFrame based on a common key.

left_on="iso3_code": This specifies the column in the gdf (the left DataFrame) to be used as the key for the merge. The iso3_code is a standard three-letter country code.

right_on="CCA3": This specifies the column in the df (the right DataFrame) to be used as the key. The CCA3 column in the CSV is assumed to also contain the three-letter country codes.

how="left": This performs a "left" join. It keeps all rows from the gdf GeoDataFrame and adds the matching data from the df DataFrame. If a match is not found, the new columns will be filled with NaN (Not a Number).

gdf.to_file("world_population_joined.shp"): This command saves the new, merged GeoDataFrame to a new Shapefile named world_population_joined.shp.

gdf.to_file("world_population_joined.gpkg", driver="GPKG"): The commented-out line shows a more modern and robust way to save the data using the GeoPackage format, which can store both the geographic data and all of its associated attributes in a single file.


print("Join complete!...") and print(gdf.head()): These lines provide confirmation that the process finished successfully and show a preview of the newly created gdf with the merged data.

```python
import geopandas as gpd
import pandas as pd

# Paths
shapefile_path = "/content/sample_data/world-administrative-boundaries-countries.shp"   # replace with your shapefile path
csv_path = "/content/sample_data/world_population.csv"

# 1. Load shapefile
gdf = gpd.read_file(shapefile_path)

# 2. Load CSV
df = pd.read_csv(csv_path)

# Print columns of df to diagnose KeyError
print("Columns in df (CSV DataFrame):")
print(df.columns)

# Drop the duplicate 'sub_region_' column from the CSV dataframe
# df = df.drop(columns=['sub_region_']) # Removed this line

# Print columns of gdf to diagnose potential duplicates
print("\nColumns in gdf (GeoDataFrame):")
print(gdf.columns)

# Drop one of the duplicate 'sub_region_' columns from gdf
gdf = gdf.loc[:,~gdf.columns.duplicated()]

# 3. Perform join on ISO3 code
#    (make sure both columns are strings and have same format)
gdf = gdf.merge(df, left_on="iso3_code", right_on="CCA3", how="left")

# 4. Save to new shapefile or GeoPackage
# gdf.to_file("world_population_joined.shp")    # ESRI Shapefile
# or better:
# gdf.to_file("world_population_joined.gpkg", driver="GPKG")

print("Join complete! New file saved.")
print(gdf.head())
```

```python
import contextily as ctx
import matplotlib.pyplot as plt

# Assuming 'gdf' from cell jAKdJ9v8jZHT contains the merged GeoDataFrame
# and 'world_pop' from cell 595VNS4pddNK contains the cluster assignments.

# Merge the 'cluster' column from world_pop onto gdf
# Ensure the merge is based on the common identifier (like 'CCA3' or 'iso3_code')
# Let's use 'CCA3' from world_pop and 'iso3_code' from gdf as the join keys
gdf_clustered = gdf.merge(world_pop[['CCA3', 'cluster']], left_on="iso3_code", right_on="CCA3", how="left")

# Convert to Web Mercator (required for contextily basemaps)
# Handle potential NaNs in the 'cluster' column that might result from the merge
gdf_clustered['cluster'] = gdf_clustered['cluster'].fillna(-1) # Assign a unique value for countries not in the cluster analysis (e.g., those dropped due to NaN values in the clustering features)
world_web = gdf_clustered.to_crs(epsg=3857)

# Plot with basemap
ax = world_web.plot(column='cluster', figsize=(12, 8), legend=True)
# Use a different provider, e.g., OpenStreetMap.Mapnik
ctx.add_basemap(ax, source=ctx.providers.OpenStreetMap.Mapnik)
plt.title("Population Density")
plt.show()
```
